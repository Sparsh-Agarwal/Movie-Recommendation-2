```python
# importing modules and libraries
import numpy as np
import pandas as pd
```

```python
# loading the dataset
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
print("Shape of movies dataset : ", movies.shape)
print("Shape of credits dataset : ", credits.shape)
```

```python
movies.rename(columns= {'id': 'movie_id'}, inplace=True)
```

```python
movies = movies.merge(credits, on=['movie_id','title'] )
```

```python
movies.columns
```

Now we are making a content based recommender system. So, choosing the right columns.

Important columns : genres, id, keywords, title, overview, cast, crew

```python
# Choosing the columns
movies = movies[['movie_id', 'title','overview', 'genres', 'keywords', 'cast', 'crew', 'tagline', 'release_date']]
```

```python
movies.head(3)
```

```python
# Checking null values
movies.isnull().sum()
```

```python
# Dropping null values
movies['overview']= movies['overview'].fillna('')
movies['tagline']= movies['tagline'].fillna('')
movies['release_date']= movies['release_date'].fillna('')
```

```python
movies['year'] = movies['release_date'].apply(lambda x: [i for i in x.split('-')]).apply(lambda x: x[0])
```

```python
movies.head()
```

```python
# Checking duplicate values
movies.duplicated().sum()
```

# Data Preprocessing

```python
movies.iloc[0].genres
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(3)
```

```python
movies.iloc[0].cast
```

```python
def convert3(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=5:
            l.append(i['name'])
            counter += 1
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
        
        elif i['job'] == 'Producer':
            l.append(i['name'])
        
        elif i['job'] == 'Writer':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies['tagline'] = movies['tagline'].apply(lambda x: x.split())
movies['title_list'] = movies['title'].apply(lambda x: x.split())
```

```python
# Removing spaces within elements in columns genres, keywords, cast, crew
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['title_list'] + movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] + movies['tagline']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'year', 'tags']]
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

# Converting Words into Vectors & Stemming

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem('dancing')
```

```python
def stemmer(text):
    y =[]
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stemmer)
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(ngram_range=(1,2), max_features=10000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
movies.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity.shape
```

```python
similarity[0]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key= lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Spider-Man')
```

```python
movie_name = input(' Enter your favourite movie name : ')

list_of_all_titles = movies['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies[movies.title == close_match].index[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies[movies.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
